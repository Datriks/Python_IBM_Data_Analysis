## Module_4 : Model Development

We can fit a linear regression using the feature 'long' - longitude and evaluate the model

```python
x = df1[['long']]
y = df1[['price']]

lm = LinearRegression()

model = lm.fit(x, y)

print('R^2 = ', model.score(x, y)*100, '%')
```

Create a new model based on te feature 'sqft_living' and 'price'

```python
# another way of transforming series into array instead of using [[]] which is easiear
x1 = df1['sqft_living'].array.reshape(-1, 1)
y1 = df1['price'].array.reshape(-1, 1)

lm1 = LinearRegression()

model1 = lm1.fit(x1, y1)

print('R^2 = ', model1.score(x1, y1)*100, '%')
```

Let's see if the model improves - value of R^2 is higher - if we use a large number of features.

```python
features = ['floors', 'waterfront', 'lat', 'bedrooms', 'sqft_basement', 'view', 'bathrooms', 'sqft_living15', 'grade', 'sqft_living']
```

```python
x2 = df1[features]
y2 = df1[['price']]

lm2 = LinearRegression()

model2 = lm2.fit(x2, y2)

R2 = model2.score(x2, y2)
print('R^2 = ', R2*100 , '%')
```

Create a polynomial model using pipe  
The first element of the tuple is the estimator  
The second element of tupple is the model constructor

```python
Input = [('scale', StandardScaler()), 
         ('polynomial', PolynomialFeatures(include_bias = False)), 
         ('model', LinearRegression())]
```

Let's use the PipeLine object to predict the price , fit the object using the feature 'features' and calculate R^2

```python
pipe = Pipeline(Input)
pipe
```

```python
model_pipe = pipe.fit(x2, y2)
```

```python
R_pipe = model_pipe.score(x2, y2)
print('R^2 = ', R_pipe * 100, '%')
```

## Model_5 : Model Evaluation and Refinement

Import the necessary modules

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
print('done')
```

Split the dta into training and testing data

```python
features = ['floors', 'waterfront', 'lat', 'bedrooms', 'sqft_basement', 'view', 'bathrooms', 'sqft_living15', 'grade', 'sqft_living']
x4 = df1[features]
y4 = df1[['price']]

x_train, x_test, y_train, y_test = train_test_split(x4, y4, test_size=0.15, random_state=1)

print('number of train_sample : ', x_train.shape[0])
print('number of test sample : ', x_test.shape[0])
```

Create and fit a Ridge Regression object using the training dataset with a regularisation parameter set to 0.1 and calculate R^2 using the test data.

```python
from sklearn.linear_model import Ridge
```

```python
Ridge1 = Ridge(alpha=0.1)

RidgeModel = Ridge1.fit(x_train, y_train)
R_ridge = RidgeModel.score(x_test, y_test)

print('R^2 = ', R_ridge * 100, '%')
```

Perform a second order polynomial transform on both the training data and testing data  
Create and fit a Ridge regression object using the training data  
Set the regularisation to 0.1  
Calculate R^2 utilising the test data provided

```python
pr = PolynomialFeatures(degree = 2)

x_train_pr = pr.fit_transform(x_train)
x_test_pr = pr.fit_transform(x_test)

Ridge2 = Ridge(alpha = 0.1)

RidgeModel2 = Ridge2.fit(x_train_pr, y_train)

R_ridge2 = RidgeModel2.score(x_test_pr, y_test)
print('R^2 = ', R_ridge2 * 100, '%')
```
