## Import the data frame

```python
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(path, names = headers)
```

## Cleaning of the data frame

```python
missing_data = df.isnull()
missing_data.head(5)
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

```python
df.replace("?", np.nan, inplace=True)
```

```python
df[["highway-mpg", "price"]].isnull().value_counts()
```

Calculate the price average and insert into nan

```python
avg_price = df["price"].astype(float).mean()
```

```python
df["price"].replace(np.nan, avg_price, inplace = True)
```

## Save the data frame

```python
df.to_csv("cleaned_data_frame.csv")
```

## Creating the model

```python
from sklearn.linear_model import LinearRegression
```

## Calculate the liniar model - Liniar Regression line

```python
lm=LinearRegression()
X = df[['highway-mpg']]
Y = df['price']
lm.fit(X, Y)  # we first train the model
Yhat=lm.predict(X)
```

```python
Yhat
```

## Calculate R^2

```python
lm = LinearRegression()
X = df[['highway-mpg']]
Y = df['price']
lm.fit(X, Y)
out=lm.score(X,Y)
```

```python
out
```
